```python
# Store the movie number of each row in list 
movie_id = None
movie_no = []

# iterate over the dataset in customer id
for cust_id in df_ratings['Customer_id']:
    if ':' in cust_id:
        # update the current movie id
        movie_id = int(cust_id.replace(":", ""))
    movie_no.append(movie_id)
```

```python
# Add new column "movie_id" in dataframe
df_ratings['Movie_id'] = movie_no
```

**Pre- filtering** :- <br/>
- now we will remove all the users that have rated less number of movies
- also all those movies that has been rated less in number

```python
# Create the df with respect to movie Id and their ratings count
movie_summmary = df_ratings.groupby('Movie_id')['Ratings'].agg(['count'])
```

```python
movie_summmary
```

Here, <br/>
1 id is having 547, means 1st movie is rated by 547 customers <br/>
2 id is having 145, means 2nd movie is rated by 145 customers <br/>
.....

```python
# Now set a benchmark for 60th percentile
movie_benchmark = round(movie_summmary['count'].quantile(0.6), 0)
movie_benchmark
```

```python
# Create a list of movie index that need to be removed from the dataset
drop_movie_list = movie_summmary[movie_summmary['count'] < movie_benchmark].index
drop_movie_list
```

```python
# Display the total number of movie needs to be removed from dataset
len(drop_movie_list)
```

```python
# Create a df with respect to customers and total number of movies they've rated
customer_summary = df_ratings.groupby('Customer_id')["Ratings"].agg(["count"])
```

```python
customer_summary
```

```python
# set a benchmark for 60th percentile
customer_benchmark = round(customer_summary['count'].quantile(0.6), 0)
customer_benchmark
```

```python
# Create a list of customer_id index that need to be removed from the dataset
drop_customer_list = customer_summary[customer_summary['count'] < customer_benchmark].index
drop_customer_list
```

```python
# Display the total number of customer needs to be removed from dataset
len(drop_customer_list)
```

```python
# Drop the cutomers & movies from the dataset
df_ratings = df_ratings[~df_ratings['Customer_id'].isin(drop_customer_list)]
df_ratings = df_ratings[~df_ratings['Movie_id'].isin(drop_movie_list)]
```

```python
# Import the libraries from surprise
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
#  SVD (Singular value Decomposition) is used for recommendation engine etc

"""
The surprise library in Python is used for building and analyzing recommender systems. It provides tools to work with collaborative
 filtering algorithms,
which are commonly used to recommend items (such as movies, books, or products) to users based on their past interactions or preferences.
"""
```

```python
# Instance of reader class
reader = Reader()
```

```python
# We only take top 100k rows for quick runtime
data = Dataset.load_from_df(df_ratings[['Customer_id','Movie_id','Rating']][ : 100000], reader)
```

```python
data
```

```python
# Create the model
model = SVD()
```

SVD as it is a recomendation engione, the output of the model i.e., <br/> 
predicted data is the estimation score on which it will decide that movie to be recomended or Not. <br/>

The estimation score will be in the range of 1-5

```python
cross_validate(model, data, measures=["RMSE"], cv=3)
```

Run a cross validation procedure for a given algorithm, reporting accuracy measures and computation times. <br/><br/>

Here, cv = 3 means, the data is splited into 3 part randomly i.e., (1, 2, 3) and then <br/>
1. First time, 1 & 2 taken as training data and 3 will taken as testing data
2. second time 2 & 3 taken as training data and 1 will taken as testing data
3. third time, 1 & 3 taken as training data and 2 will taken as testing data <br/>
and perform the training & testing, and computes the metrics for particular part and shows it.

**Note** :- <br/>
This method makes sure that each sub-part data has been taken as training and testing once. And shows the performance of each data
